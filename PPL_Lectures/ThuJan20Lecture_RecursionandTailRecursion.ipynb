```scala
def factorial (x: Int): Int= {
    if (x<= 1) {
        1
    }
    else {
        x*factorial(x-1)
    }
}
```

```scala
//1,1,3,5,13,21
def fibonacci(x:Int): Int={
    if (x<=2) {
        1 //basecase 
    }
    else {
        fibonacci(x-1)+ fibonacci(x-2) //neither are tail calls 
    }
}
```

```scala
fibonacci(10)
```

```scala
def foo(x:Int): Int ={
    if(x<=1) 1
    else{
        foo(x-1) //this is called a tail call because whatever this call
                //returns is what returns back to the caller unchanged
    }
}
//always returns 1 for any input
```

```scala
foo(8)
```

```scala
def factorialTail(x:Long, acc:Long): Long = {
    if (x<=1){
        acc
    }
    else {
        factorialTail(x-1, acc*x)
    }
}
//this is tail recursive! we are premultiplying before we ever return
//normal factorial multiplies after we return 

def myTailFactorial(x:Long): Long={
    factorialTail(x,1) //initialize acc=1
}
```

```scala
myTailFactorial(32000) //no stack overflow because tail 
```

```scala
factorial(32000) //stack overflow
```

```scala
import scala.annotation.tailrec
```

```scala
//cannot use @tailrec here
def bar(x:Long): Int ={
    if(x<=1) 1
    else{
        bar(x-1)+2 //this is NOT a tail call because there are opperations performed on the return value of the recursive call
    }
}
//NOT tail becasue more work left to be done after the return value
```

```scala
@tailrec //you have to import scala.annotation.tailrec then use @tail rec for scala to know
def barTail(x:Long, acc:Int): Int ={
    if (x<=1){
        acc
    }
    else{
        barTail(x-1,acc+2) //preadd instead of post add makes it tail recursive
    }
}
def myBarTail(x:Long):Int={
    barTail(x,1)
}
```

```scala
myBarTail(4)
```

```scala
myBarTail(1000000000L) //since we are using tail recursive this will work
```

```scala
bar(1000000000L) //stack overflow occurs becasue we are not using tail recursion
```
